```python
class Solution:
    def maxVowels(self, s: str, k: int) -> int:
        vowels=set("aeiou")
        count = maxCount=0

        for i in s[0:k]:
            if i in vowels:
                count += 1
        maxCount = count

        for i in range(k,len(s)):
            if s[i-k] in vowels:
                count -= 1
            if s[i] in vowels:
                count += 1
            if maxCount < count:
                maxCount = count
        return maxCount
```

```python
# words = ["listen", "silent", "enlist", "bad", "dab", "ab", "bca"]

from collections import Counter


def anagram_groups1(words):
    l = []
    o = []
    for word in words:
        c = Counter(word)
        l.append((c, word))
    l.sort(key=lambda x: x[0])
    # print(len(l))
    p = l[0][0]
    sub_list = []
    for c, word in l:
        if c == p:
            sub_list.append(word)
        else:
            o.append(sub_list)
            sub_list = [word]
        p = c
    if len(sub_list) == 1:
        o.append(sub_list)
    return o

anagram_groups1(words=["listen", "silent", "enlist", "bad", "dab", "ab", "bca"])
```

```python
# words = ["listen", "silent", "enlist", "bad", "dab", "ab", "bca"]

from collections import Counter, defaultdict


def anagram_groups2(words):
    l = defaultdict(list)
    for word in words:
        c = frozenset(Counter(word))
        l[c].append(word)
    return list(l.values())

anagram_groups2(words=["listen", "silent", "enlist", "bad", "dab", "ab", "bca"])
```

```python
class Stock:

    def __init__(self) -> None:
        self._price = None

    @property
    def price(self):
        if self._price:
            return self._price
        else:
            self._price = 20
    
    @price.setter
    def price(self, val):
        if val <= 0:
            raise ValueError("value should always be positive")
        self._price = val

    

s = Stock()
```
