```python
import numpy as np
import pandas as pd


def solution(files):
    # files - any of available files, i.e:
    # files = ["./data/framp.csv", "./data/gnyned.csv", "./data/gwoomed.csv",
    #            "./data/hoilled.csv", "./data/plent.csv", "./data/throwsh.csv",
    #            "./data/twerche.csv", "./data/veeme.csv"]

    # write your solution here
    output = []
    for file in files:
        df = pd.read_csv(file, usecols=["date", "vol", "close"])
        df["year"] = df["date"].apply(lambda x: x.split("-")[0])
        df["date"] = pd.to_datetime(df["date"])
        # df_max = df[["year", "vol", "close"]].groupby(["year"]).max()
        df_vol = pd.merge(
            df[["date", "vol", "year"]],
            df[["year", "vol"]].groupby(["year"]).max(),
            on=["year", "vol"]
        ).reset_index()
        df_close = pd.merge(
            df[["date", "close", "year"]],
            df[["year", "close"]].groupby(["year"]).max(),
            on=["year", "close"]
        ).reset_index()
        output.append([df_vol, df_close])
    return output


```

```python
def solution(blocks):
    lowest_point = min(blocks)
    lowest_position = blocks.index(lowest_point)
    cur_position = blocks[0]
    starting_position = blocks[0]
    for e in blocks:
        if e >= lowest_point and e >= cur_position:
            cur_position = e
        if e < lowest_point and cur_position > lowest_point:
            return cur_position - starting_position + 1
    
```

```python
def solution(blocks):
    start = 0
    end = 1
    prev = blocks[0]
    curr = blocks[1]
    diff = 2
    last = len(blocks) - 1
    if last == 1:
        return diff
    for end, next in enumerate(blocks[2:], 2):
        # print(diff, start, end, prev, curr, next)
        if (curr > prev and curr > next):
            if end - start > diff:
                diff = end - start
            start = end - 1
        prev = curr
        curr = next
    # print(diff, start, end, prev, curr, next)
    if end == last:
        if end - start + 1 > diff:
            diff = end - start + 1
    return diff

print(solution([2, 6, 8, 5]))
print(solution([2, 2]))
print(solution([1, 6, 5, 2, 4]))
    
```
